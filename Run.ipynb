# Setting Up 

```python
import os
import cv2
import numpy as np

from tensorflow.keras.models import load_model
from tensorflow.keras.applications import vgg16

from tf_pose.estimator import TfPoseEstimator
from tf_pose.networks import get_graph_path, model_wh

%matplotlib inline
```

```python
saved_models_path = "saved_models"
vgg16nn_path = os.path.join(saved_models_path,"finetuned_VGG16nn.h5")
```

```python
video = "vid.mp4"
```

# Creating Folder to save the frames

**The frames will have their labels and index mentioned in their filename itself**

```python
vid2frames_path = "vid2frames"
if not os.path.exists(vid2frames_path):
    os.makedirs(vid2frames_path)
```

# Loading up the Models

**VGG16NN**

```python
vgg16nn = load_model(vgg16nn_path)
```

**OpenPose**

```python
model = "mobilenet_v2_large"
resize = "432x368"
resize_out_ratio = 4.0
w, h = model_wh(resize)
if w == 0 or h == 0:
    e = TfPoseEstimator(get_graph_path(model), target_size=(432, 368))
else:
    e = TfPoseEstimator(get_graph_path(model), target_size=(w, h))
```

# Running on Video

```python
cap = cv2.VideoCapture(video)
c=0
while True:
    ret, frame = cap.read()
    c = c+1
    if ret == False:
        break
    humans = e.inference(frame, resize_to_default=(w > 0 and h > 0), upsample_size=resize_out_ratio)
    image_with_pose_pts = TfPoseEstimator.draw_humans(frame, humans, imgcopy=False)  
    pp_array = np.array([0 for i in range(108)])
    for h, human in enumerate(humans[:2]):
        for id,body_part in human.body_parts.items():
            pos = 3 * id
            #print(pos)
            pp_array[pos + (h*54)] = body_part.x
            pp_array[pos + 1 + (h*54)] = body_part.y
            pp_array[pos + 2 + (h*54)] = body_part.score

    resized = cv2.resize(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB), (224,224), cv2.INTER_NEAREST)
    preprocessed_frame = vgg16.preprocess_input(resized)
    preprocessed_frame = np.expand_dims(preprocessed_frame, axis = 0)
    pp_array = np.expand_dims(pp_array, axis = 0)
    pred = vgg16nn.predict([preprocessed_frame, pp_array])
    if pred <= 0.5:
        label = "Fighting"
    else:
        label = "Not_Fighting"
    cv2.imwrite(os.path.join(vid2frames_path,"{0}_{1}.jpg".format(c, label)), image_with_pose_pts)
```
